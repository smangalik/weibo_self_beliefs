# Extract content from the Leidin Weibo Corpus

```python
import pandas as pd
import numpy as np
import os
from tqdm.notebook import tqdm
import warnings
warnings.filterwarnings("ignore")

print("Reading JSON files from 'weibo2' directory...")
df = []
for file in tqdm(os.listdir('weibo2')):
    if file.endswith('.json'):
        json_df = pd.read_json(os.path.join('weibo2', file), encoding='utf-8', lines=False)
        json_df['repost?'] = json_df['repost'].apply(lambda x: 0 if x in [None, np.nan, "NaN"] else 1)
        json_df[['weibo_id','user_id','content','timestamp','repost?']]
        df.append(json_df[['weibo_id', 'user_id', 'content', 'timestamp', 'repost?']])
        
df = pd.concat(df, ignore_index=True)
df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s', utc=True)
df = df.sort_values(by='timestamp', ascending=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df[df['repost?'] == 0].to_csv('leidin_weibo_corpus_no_reposts.csv', index=False, encoding='utf-8')
df[df['repost?'] == 1].to_csv('leidin_weibo_corpus_reposts.csv', index=False, encoding='utf-8')
```
